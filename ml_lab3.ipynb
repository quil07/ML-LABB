#Linear regression


with out data set

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([1,2,3,4,5])
y = np.array([3,4,2,4,5])
x_mean = np.mean(x)
y_mean = np.mean(y)
numerator = np.sum((x-x_mean)*(y-y_mean))
denominator = np.sum((x-x_mean)**2)
m = numerator / denominator #slope
b = y_mean - (m*x_mean) #intercept
print("slope - ",m)
print("Intercept -" ,b)
#prediciton
y_pred = m*x + b
print("Prediction - ",y_pred)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression")
plt.plot(x,y_pred,color='red')
plt.show()
```

#values in array formatusing skitkit learn

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
x = np.array([[0],[1],[2],[3],[4]])
y = np.array([2,3,4,5,6])
model = LinearRegression()
model.fit(x,y)
print("slope : ",model.coef_[0])
print("intercept : ",model.intercept_)
predictions = model.predict(x)
print("prediction : ",predictions)
plt.scatter(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression")
plt.plot(x,model.predict(x),color='red')
plt.show()
```

#mse and rmse


```python
import numpy as np
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
#sample data (replace your actual true and predicted value)
y_true = [3,-0.5,2,7]
y_pred = [2.5,0.0,2,8]
#calculate the matrices
mse = mean_squared_error(y_true,y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_true,y_pred)
r2 = r2_score(y_true,y_pred)
#print the result
print("Mean Squared Error (MSE) : ",mse)
print("Root Mean Squared Error (RMSE) : ",rmse)
print("Mean Absolute Error (MAE) : ",mae)
print("R-squared (R2) Score : ",r2)
```

#simple linear regression

```python
from typing import ValuesView
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

#importing dataset

```python

datase =  pd.read_csv('salary.csv')
x = datase.iloc[:, :-1].values
y = datase.iloc[:,-1].values
```

#splitting the dataset into trainning and testing set

```python

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

#training the simple linear rigression model on the traininng set


```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

#predicting the test set result

```python
y_pred = regressor.predict(x_test)
```

#Visulaizing thre trainning set result

```python
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,regressor.predict(x_train),color='blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
years_experience = float(input("Enter Years of Experience to predict salary: "))
# Reshape the input into a 2D NumPy array as expected by the predict method
experienced_years_reshaped = np.array([[years_experience]])

# Predict the salary for the given years of experience
predicted_salary = regressor.predict(experienced_years_reshaped)

# Display the predicted salary
print(f"For {years_experience} years of experience, the predicted salary is: ${predicted_salary[0]:.2f}")
```
