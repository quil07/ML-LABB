3. MEDIAN

```python
data = [10, 20, 30, 40, 50]
data.sort()
n=len(data)
if n%2==0:
  median = data[n//2-1] + data[n//2] / 2
else:
  median = data[n//2]
print("Median:",median)
```

4. MODE

```python
from collections import Counter

data = [10, 20, 30, 40, 50, 20, 30, 20]
frequency = {}
for i in data:
  if i in frequency:
    frequency[i] += 1
  else:
    frequency[i] = 1
mode = max(frequency, key=frequency.get)
print("Mode:", mode)
```

5. STANDARD DEVIATION

```python
import math
data = [10, 20, 30, 40, 50]
mean = sum(data) / len(data)
print(f"Data: {data}")
print(f"Mean: {mean}")
squared_diffs = [(x - mean) ** 2 for x in data]
print(f"Squared Differences: {squared_diffs}")
variance_population = sum(squared_diffs) / len(data)
print(f"Variance (Population): {variance_population:.2f}")
std_dev_population = math.sqrt(variance_population)
print(f"Standard Deviation (Population, manual): {std_dev_population:.2f}")
```
